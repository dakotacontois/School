```python
#importing packages and identifying/importing csv file

import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import association_rules, apriori
from mlxtend.preprocessing import TransactionEncoder
medical=pd.read_csv('medical_market_basket.csv')
```

```python
# remove empty rows
medical=medical[medical['Presc01'].notna()]
medical.head(10)

#empty rows removed
```

```python
medical.shape
```

```python
#reset index so index wont show 15000 rows
medical.reset_index(drop=True, inplace=True)
medical.info()
```

```python
# create list of lists

transac_list = []

for index, row in medical.iterrows():
    presc = []
    for col in row:
        if not pd.isna(col):
            presc.append(col)
    transac_list.append(presc)

# show first and last rows of the list
print(f"First line:", transac_list[0], 
      "Last line:", transac_list[7500])
```

```python
# load in transaction encoder and transform data
TE=TransactionEncoder()
array=TE.fit(transac_list).transform(transac_list)

transaction=pd.DataFrame(array,columns=TE.columns_)
transaction
```

```python
# bring in apriori and apply to dataset
freq_items = apriori(transaction, min_support = 0.02, use_colnames = True)
freq_items
```

```python
# set lift threshold. Need to use num_itemsets for most recent software update; set as 2 
rule_table= association_rules(freq_items, metric = 'lift', min_threshold=1, num_itemsets=2)
rule_table
```

```python
# top 3 rules for confidence
top_3_rules_conf=rule_table.sort_values('confidence', ascending=False).head(3)
top_3_rules_conf
```

```python
# top 3 rules for support
top_3_rules_supp=rule_table.sort_values('support', ascending=False).head(3)
top_3_rules_supp
```

```python
# top 3 rules for lift
top_3_rules_lift=rule_table.sort_values('lift', ascending=False).head(3)
top_3_rules_lift
```

```python
# top 3 overall rules using a lift > 1.5, confidence >0.4, support defined earlier as >0.02
# want a minimum lift to indicate a posiitve relationship
# sorting by confidence desc because confidence is the probability of the consequent happening in combination of the antecedent
top_3_rules = rule_table[(rule_table['lift'] > 1.5) & 
                         (rule_table['confidence'] > 0.4)].sort_values(by=['confidence'], 
                                                                       ascending= False).head(3)
top_3_rules

#interesting that abilify is the consequent of all three and the top 3 rules for confidence are the same as here
```
