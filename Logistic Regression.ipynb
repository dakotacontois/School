```python
# Select independent variables
X = medical[['Children', 'Age', 'Income', 'VitD_levels','Soft_drink_numeric','Diabetes_numeric','Stroke_numeric',
             'Complication_risk_numeric','Overweight_numeric','Initial_days','Emergency_numeric','Elective_numeric',
             'CTScan_numeric','Intravenous_numeric','MRI_numeric']]

# Calculate VIF for each independent variable
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["features"] = X.columns

# Print the VIF results
print(vif)
```

```python
#remove VitD levels due to high VIF (multicollinearity)
# this is the initial model
x = medical[['Children', 'Age', 'Income','Diabetes_numeric','Stroke_numeric', 'Soft_drink_numeric',
             'Complication_risk_numeric','Overweight_numeric','Initial_days','Emergency_numeric','Elective_numeric',
             'CTScan_numeric','Intravenous_numeric','MRI_numeric']]
y = medical['ReAdmis_numeric']

x = sm.add_constant(x) # adding a constant

initial_model = sm.Logit(y, x).fit()
 
print(initial_model.summary())
```

```python
# create confusion matrix for initial to final model comparison/evaluation metric
conf_matrix=initial_model.pred_table()

print(conf_matrix)
```

```python
# Test accuracy of initial matrix
TN=conf_matrix[0,0]
TP=conf_matrix[1,1]
FN=conf_matrix[1,0]
FP=conf_matrix[0,1]

accuracy = (TN + TP) / (TN + FN + FP + TP)

print(accuracy)
```
